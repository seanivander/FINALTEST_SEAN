```python
import json
import requests
import pandas as pd
```

```python
# Request API

response = requests.get('https://berita-indo-api-next.vercel.app/api/kumparan-news')
data = response.json()
```

```python
# Ubah JSON response api ke dataframe

data_json = json.dumps(data['data'])
df = pd.read_json(data_json)
df.head()
```

```python
# Transformasi data kolom 'isoDate' menjadi tipe data datetime

df['isoDate'] = pd.to_datetime(df['isoDate'])
df.info()
```

```python
# Aggregasi data berdasarkan kolom 'creator', serta urutkan berdasarkan jumlah berita terbanyak

df.groupby('creator').size().reset_index(name='count').sort_values(by='count', ascending=False)
```
